```python
gold = pd.read_csv('../data/minigold.csv',delimiter='\t')
system = pd.read_csv('../data/miniout1.csv',delimiter='\t')
gold_annotations = gold['gold']
system_annotations = system['NER']
categories = set(gold_annotations)

def get_true_positives(tag,gold,system):
    counter = 0
    for pair in zip(gold,system):
        if pair[0] == tag and pair[1] == tag:
            counter += 1
    return counter

def get_true_positives(tag,gold,system):
    counter = 0
    for pair in zip(gold,system):
        if pair[0] == tag and pair[1] == tag:
            counter += 1
    return counter

def get_false_negatives(tag,gold,system):
    counter = 0
    for pair in zip(gold,system):
        if pair[0] == tag and pair[1] != tag:
            counter += 1
    return counter

def get_false_positives(tag,gold,system):
    counter = 0
    for pair in zip(gold,system):
        if pair[0] != tag and pair[1] == tag:
            counter += 1
    return counter

def get_true_negatives(tag,gold,system):
    counter = 0
    for pair in zip(gold,system):
        if pair[0] != tag and pair[1] != tag:
            counter += 1
    return counter

def get_precision(tag,gold,system):
    true_positives = get_true_positives(tag,gold,system)
    false_positives = get_false_positives(tag,gold,system)
    n_system = true_positives + false_positives
    try:
        precision = true_positives / n_system
    except ZeroDivisionError:
        precision = 0.0
    return precision

def get_recall(tag,gold,system):
    true_positives = get_true_positives(tag,gold,system)
    false_negatives = get_false_negatives(tag,gold,system)
    n_gold = true_positives + false_negatives
    recall = true_positives / n_gold
    return recall

def get_f_score(tag,gold,system):
    precision = get_precision(tag,gold,system)
    recall = get_recall(tag,gold,system)
    f_score_numerator = 2*precision*recall
    f_score_denominator = precision+recall
    try:
        f_score = f_score_numerator / f_score_denominator
    except ZeroDivisionError:
        f_score = 0.0
    return f_score

def get_precision_recall_f_score(tag,gold,system):
    scores = {}
    scores["precision"] = get_precision(tag,gold,system)
    scores["recall"] = get_recall(tag,gold,system)
    scores["f_score"] = get_f_score(tag,gold,system)
    return scores

matrix = confusion_matrix(gold_annotations, system_annotations)
print(matrix)

for category in categories:
    print(category)
    print("scores",get_precision_recall_f_score(category,gold_annotations,system_annotations)) 
```
